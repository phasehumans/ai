```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import skew
```

```python
# data creation

np.random.seed(42)
# right skewed distribution
income_distribution= np.random.gamma(shape=2.0, scale=1000, size=1000)

skew_value= skew(income_distribution)

sns.histplot(income_distribution, bins=30, kde= True)
plt.title(f"Income distribution(skewness: {skew_value:.2f})")
plt.xlabel('Income')
plt.ylabel('Frequency')
```

```python
print(data.shape)
print(data.size)
```

```python
# range = max - min 

temp_range= data['temperature'].max() - data['temperature'].min()
print(f"Range => {temp_range: .3f}")
```

```python
# IQR (Q3 - Q1)

Q1= data['temperature'].quantile(0.25)
Q3= data['temperature'].quantile(0.75)

IQR= Q3-Q1
print(IQR)
```

```python
# variance of pressure col

pressure_var= data['pressure'].var()
print(pressure_var)
```

```python
# standard deviation

pressure_std= data['pressure'].std()
print(pressure_std)
```
