```python
# data creation in nominal

import numpy as np
import pandas as pd

np.random.seed(0)
# for consistent data (same res)

number_samples= 1000
clrs= ['red', 'blue', 'green', 'purple', 'yellow', 'orange']

nominal_data= np.random.choice(clrs, number_samples)
# choice when structure data - clrs

nominal_data
```

```python
# convert into dframe
data= pd.DataFrame(nominal_data,columns= ['Colors'] )

data
```

```python
# data creation in ordinal 

satisfaction_lvl= ['low', 'medium', 'high']

ordinal_data= np.random.choice(satisfaction_lvl, number_samples)

# data= pd.DataFrame(ordinal_data, columns=['Satisfaction Level']) --> for creating new dataframe

data['Satisfaction Level'] = ordinal_data
data
```

```python
# data creation interval data

np.random.seed(0)
temp= np.random.randint(0,100)
temp2= np.random.randint(0,100, number_samples)

interval_data= np.random.choice(temp, number_samples)
data['Temperature']= interval_data
data['Temperature 2']= temp2

data
```

```python
# data creation of range data

np.random.seed(0)
income= np.random.randint(10000, 250000, number_samples)

data['income']= income
data
```

```python
data= data.drop('Temperature 2', axis= 1) # axis 1 col, 0= row

data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.hist(data['income'], bins=20, edgecolor= 'black')
plt.title('Income distribution')
plt.xlabel('income in rupess')
plt.ylabel('frequency')

# plt.savefig('02_income_dist.png')
plt.show()
```
