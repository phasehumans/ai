```python
# simple random sampling

sample_data= population.sample(n= 100, random_state= 44)  #random state -> same (seed)

sample_data
```

```python
# Stratified random sampling

startifeid_data= population.groupby('Region').apply(lambda x: x.sample(min(len(x),25), random_state= 44))
startifeid_data.to_csv('stratified_data.csv')

print(startifeid_data)
```

```python
# cluster sampling

clusters= manufacture['BatchNumber'].unique()
print(clusters)

selected_clusters= np.random.choice(clusters, size= 5, replace=False)
print(selected_clusters)

cluster_data= manufacture[manufacture['BatchNumber'].isin(selected_clusters)]
print(cluster_data)
```
